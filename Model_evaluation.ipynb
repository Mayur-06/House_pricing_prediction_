```python
import joblib

# Load the trained model saved in previous file (Model_training.ipynb)
model = joblib.load('Dataset/model.pkl')

# Now you can use the loaded model for predictions in this notebook
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
# Load the datasets from files which were previously saved (Data_splitting.ipynb)
X_train = pd.read_csv('Dataset/X_train.csv')
X_test = pd.read_csv('Dataset/X_test.csv')
Y_train = pd.read_csv('Dataset/Y_train.csv')
Y_test = pd.read_csv('Dataset/Y_test.csv')


# Now you have the datasets ready to use in this notebook
#actually you don't need to create csv files for everything and load and dumb from one notebook to another....
#I have done this to just separately show the process and steps in ML model building. You can run the whole code of project in one notebook also
```

Prediction of training data

```python
#First let's predict and check how model is performing
training_data_pred = model.predict(X_train)
```

```python
print(training_data_pred)
```

Evaluation
1. R squared error
2. Mean absolute error

```python
score_1 = metrics.r2_score(Y_train, training_data_pred)
```

```python
score_2=metrics.mean_absolute_error(Y_train,training_data_pred)
```

```python
print("R squared error:", score_1)
print("Mean absolute error:", score_2)

```

In summary, based on these metrics:

Your model explains about 94.6% of the variance in the target variable.
On average, your predictions are off by about 0.191 units from the actual values.

```python
#visualizing the actual price and predicted prices
plt.scatter(Y_train, training_data_pred)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("Actual price vs predicted price")
plt.show()
```

```python
#therefore through graph we can see that value predicted by our model is very much similar to the actual prices data
#since data points are very close to each other
```
